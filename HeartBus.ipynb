```python
# 패키지 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 보호대상아동현황 데이터 불러오기
kids = pd.read_csv('D:\\Teenager\\1321_보호대상아동현황.csv', encoding='cp949')
kids.head(10)
```

```python
kids.shape
```

```python
# 결측치가 들어간 column 모두 제거
kids.dropna(axis=1, inplace=True)
```

```python
# 2013, 2013.1, 2013.2......2019.6 과 그 이후 데이터 분리
kids1 = kids.iloc[:, 0:50]
kids2 = kids.iloc[:, 51:]
```

```python
# 데이터 확인
kids1.head(10)
```

```python
# 1행, 2행의 내용을 합쳐서 index의 글자와 합치기
kids1.columns = kids1.iloc[0] + kids1.iloc[1]
```

```python
# 1행, 2행, 3행, 4행 삭제
kids1.drop([0, 1, 2, 3], inplace=True)
```

```python
kids1
```

```python
# 1행을 맨 뒤로 보내기 ('계')
kids1 = kids1[1:].append(kids1.iloc[0])
```

```python
# index의 1열을 '지역'으로 변경
kids1.columns.values[0] = '지역'
```

```python
# '지역' column을 index로 변경
kids1.set_index('지역', inplace=True)
```

```python
# kids1의 column 이름을 숫자로 변경
kids1.columns = range(1, 50)
```

```python
# kids1의 column 이름을 str로 변경
kids1.columns = kids1.columns.map(str)
```

```python
kids1.columns
```

```python
kids1.info()
```

```python
# 모든 데이터 중 index, column 제외 전부 정수형으로 전환
kids1 = kids1.astype(int)
```

```python
# kids1의 column명 중 1, 8, 15, 22, 29, 36, 43의 값을 합친 '소계' 생성
kids1['소계'] = kids1['1'] + kids1['8'] + kids1['15'] + kids1['22'] + kids1['29'] + kids1['36'] + kids1['43']
```

```python
# kids1의 column명 중 2, 9, 16, 23, 30, 37, 44의 값을 합친 '비행_가출_부랑' 생성
kids1['비행_가출_부랑'] = kids1['2'] + kids1['9'] + kids1['16'] + kids1['23'] + kids1['30'] + kids1['37'] + kids1['44']
```

```python
# kids1의 column명 중 3, 10, 17, 24, 31, 38, 45의 값을 합친 '학대' 생성
kids1['학대'] = kids1['3'] + kids1['10'] + kids1['17'] + kids1['24'] + kids1['31'] + kids1['38'] + kids1['45']
```

```python
# kids1의 column명 중 4, 11, 18, 25, 32, 39, 46의 값을 합친 '부모빈곤_실직' 생성
kids1['부모빈곤_실직'] = kids1['4'] + kids1['11'] + kids1['18'] + kids1['25'] + kids1['32'] + kids1['39'] + kids1['46']
```

```python
# kids1의 column명 중 5, 12, 19, 26, 33, 40, 47의 값을 합친 '부모사망' 생성
kids1['부모사망'] = kids1['5'] + kids1['12'] + kids1['19'] + kids1['26'] + kids1['33'] + kids1['40'] + kids1['47']
```

```python
# kids1의 column명 중 6, 13, 20, 27, 34, 41, 48의 값을 합친 '부모질병' 생성
kids1['부모질병'] = kids1['6'] + kids1['13'] + kids1['20'] + kids1['27'] + kids1['34'] + kids1['41'] + kids1['48']
```

```python
# kids1의 column명 중 6, 13, 20, 27, 34, 41, 48의 값을 합친 '부모이혼등' 생성
kids1['부모이혼등'] = kids1['7'] + kids1['14'] + kids1['21'] + kids1['28'] + kids1['35'] + kids1['42'] + kids1['49']
```

```python
# kids1의 1열부터 49열까지 제거
kids1.drop(kids1.iloc[:, 0:49], axis=1, inplace=True)
```

```python
# kids1 저장
kids1.to_csv('D:\\Teenager\\kids1.csv', encoding='cp949')
```

```python
# kids1의 '계' 행만 다른 데이터프레임으로 저장
kids1_total = kids1.iloc[0]
```

```python
# kids1_total의 '소계' 제거
kids1_total.drop('소계', inplace=True)
```

```python
# kids1_total 파이차트로 시각화
ax = kids1_total.plot.pie(figsize=(10, 10))
ax.axis('equal')
ax.legend(loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))
ax.set(ylabel='')
plt.title('청소년 가출 원인', fontsize=20)
plt.show()
```

```python
# counsel1의 '초등학생', '중학생', '고등학생' 행의 값을 더한 '미성년자' 행 생성
counsel1.loc['미성년자'] = counsel1.loc['초등학생'] + counsel1.loc['중학생'] + counsel1.loc['고등학생']
```
